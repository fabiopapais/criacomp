```python
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage
from langchain.prompts.chat import ChatPromptTemplate

def evaluator(input):
  llm = ChatOpenAI(model="gpt-4-1106-preview")

  template = "You are a helpful assistant that classify whether the sentence mentions any animal. Only respond with 'Yes' or 'No'."
  human_template = "{text}"

  chat_prompt = ChatPromptTemplate.from_messages([
    ("system", template),
    ("human", human_template),
  ])

  messages = chat_prompt.format_messages(text=input)
  response = llm.invoke(messages)
  return response.content

[evaluator("My cat is a fluffy devil"), evaluator("My car is broken")]
```

```python
import pandas as pd

sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "I had pasta for dinner last night.",
    "Birds fly south for the winter.",
    "The sun sets in the west.",
    "Cats are known for their curiosity.",
    "I need to charge my phone.",
    "A group of fish is called a school.",
    "It's raining outside.",
    "Dogs are loyal companions.",
    "I forgot to bring my umbrella."
]

labels = [
    'Yes',
    'No',
    'Yes',
    'No',
    'Yes',
    'No',
    'Yes',
    'No',
    'Yes',
    'No'
]

df = pd.DataFrame({'Sentence': sentences, 'MentionsAnimals_GroundTruth': labels})
df.head()
```

```python
df['MentionsAnimals_LLM'] = df['Sentence'].apply(evaluator)
df
```
